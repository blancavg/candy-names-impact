# Candy Candy – Did a cartoon influence name trends?

This notebook explores whether the anime *Candy Candy*, broadcast in Spain in the 1980s, contributed to a decline in the use of the name “Susana.” It compares name frequency data in Spain and Uruguay (a country without exposure to the anime) to identify any cultural influence through media.


## What do we want to know?

> Can a fictional story change how people perceive a name?

Susana’s role in *Candy Candy* left a lasting impression. We explore whether this contributed to a decline in the name's popularity in Spain, compared to Uruguay.


## 📂 Load and prepare data

```
import pandas as pd

# Load preprocessed datasets
esp = pd.read_csv('./outputs/tables/frecuencia_relativa_espana.csv')
uru = pd.read_csv('./outputs/tables//frecuencia_relativa_uruguay.csv')

esp.head(), uru.head()
```

## 📊 Relative frequency in Spain

```
import matplotlib.pyplot as plt

esp.set_index('Década').plot(marker='o')
plt.title('Relative frequency by decade – Spain')
plt.ylabel('Proportion')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 📊 Relative frequency in Uruguay

```
uru.set_index('Década').plot(marker='o')
plt.title('Relative frequency by decade – Uruguay')
plt.ylabel('Proportion')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 📉 Comparative insights

- **Susana** shows an **abrupt decline** in Spain, from 30% to below 5% between 1970–1990.
- In Uruguay, Susana had a **low initial frequency** (~3%) and decreased gradually.
- **María** consistently increased and dominated in both countries.
- Other names like **Patricia** and **Ana** show **moderate or steady declines**.

These patterns suggest a potential cultural effect related to *Candy Candy* in Spain.


## Conclusions

This exploratory analysis reveals a uniquely sharp decline in the name **Susana** in Spain during the post-*Candy Candy* era, whereas other countries without exposure, such as Uruguay, show no such pattern. Although not conclusive, the difference suggests that the media may subtly influence naming choices.
